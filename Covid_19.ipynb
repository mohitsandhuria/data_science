```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('covid_19_india.csv')
data.head()
data['State/UnionTerritory']=data['State/UnionTerritory'].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['State_label']=le.fit_transform(data['State/UnionTerritory'])
data.head()

data_dict=dict(zip(data['State/UnionTerritory'].unique(),data['State_label'].unique()))
data_dict
```

```python
data_new= data[data['Date']=="28/07/20"]
data_new['Confirmed'].sum()
data_new=data_new.sort_values(by=['Confirmed'],ascending=False)
plt.figure(figsize=(7,4), dpi=70)
plt.bar(data_new['State/UnionTerritory'][0:5],data_new['Confirmed'][0:5],)
plt.xlabel('State')
plt.ylabel('confirmed')
plt.show()
```

```python
data_new['Deaths'].sum()
data_new=data_new.sort_values(by='Deaths',ascending=False)
plt.bar(data_new['State/UnionTerritory'][0:5],data_new['Deaths'][:5])
plt.xlabel('State/UnionTerritory')
plt.ylabel('Deaths')
plt.show()
```

```python
data_new['Death/confirmed Cases']=data['Confirmed']/data['Deaths'].round(2)
data_new['Death/confirmed Cases']=[np.nan if x==float('inf') else x for x in data_new['Death/confirmed Cases']]

data_new=data_new.sort_values(by='Death/confirmed Cases',ascending=True)
data_new
```

```python
data_new=data_new.drop(['Sno','Date','Time','ConfirmedIndianNational','ConfirmedForeignNational','Unnamed: 9'], axis = 1)
data_new

```
